```python
# Write a function to find moving average in an array over a window:
# Test it over [3, 5, 7, 2, 8, 10, 11, 65, 72, 81, 99, 100, 150] and window of 3.

def movingaverage(l,k):
    N=len(l)
    y=[]
    for i in range(N-k+1):
        j=0
        s=0
        while j<k:
            s=s+l[i+j]
            j=j+1
        a=s/k
        y.append(a)
    return y

li=[3, 5, 7, 2, 8, 10, 11, 65, 72, 81, 99, 100, 150]
print(movingaverage(li,3))
```

```python
# For each value in integer column 'X' of DataFrame df, count the difference back to the previous zero (or the start of the Series, whichever is closer) and store it in a new column 'Y'

def distance(l):
    y=[]
    i=0
    for num in l:
        if num!=0:
            i=i+1
        else: 
            i=0
        y.append(i)
    return y


import pandas as pd
df1 = pd.DataFrame({'X': [7, 2, 0, 3, 4, 2, 5, 0, 3, 4]})
li=df1['X']
y=distance(li)
df2 = pd.DataFrame({'Y': y})
df = pd.concat([df1,df2], axis=1)
df
```

```python
# Create a DatetimeIndex that contains each business day of 2015 and use it to index a Series of random numbers.

import numpy as np

d= pd.bdate_range('2015-01-01', '2015-12-31')
print(d)
s = pd.Series(np.random.randint(100,size=len(d)), index=d)
print(s)
```

```python
# Find the sum of the values in s for every Wednesday.

t=s.asfreq('W-WED')
print(t)
t.sum()
```
